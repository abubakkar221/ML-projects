# Hyper parameter tunning using GridSearchCV

```python
import pandas as pd
```

```python
from sklearn import datasets
iris=datasets.load_iris()
iris
```

```python
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df
```

```python
df["flower_name"]=iris.target
df
```

```python
df["flower_name"]=iris.target_names[iris.target]
df
### 0,1,2 ar jaigai flower name bosaisi setosa,,versicolur,verginica
```

```python
x=df.drop("flower_name",axis=1)
x
```

```python
y=df["flower_name"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=6)
```

```python
x_train
```

```python
x_test
```

```python
x_test.shape
```

```python
y_train
```

```python
y_test
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
###akn amra nise grid search cv library deye dakhbo svc() ar modde kon perameter best hoi,,
###pore aisa seita debo
```

# use GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
arman=GridSearchCV(SVC(gamma="auto"),{
    "C":[1,10,20,30],
    "kernel":["rbf","linear","poly"]
}, cv=5)
arman.fit(x,y)
arman.cv_results_
```

```python
df=pd.DataFrame(arman.cv_results_)
df
```

```python
arman.best_params_
```

```python
arman.best_score_
```

# approch 2 : use k fold cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
sv1=cross_val_score(SVC(kernel="rbf",C=1,gamma="auto"),x,y,cv=5)
sv1
```

```python
sv1.mean()
```

```python
sv2=cross_val_score(SVC(kernel="linear",C=20,gamma="auto"),x,y,cv=5)
sv2
###akane amra nij theke perameter deye daksi
```

```python
sv2.mean()
###nijer issa moto kernel perameter dele accuracy kome jai
```

```python
###akn amra bivinno model deye dakhbo
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
sv3=cross_val_score(DecisionTreeClassifier(),x,y,cv=5)
sv3
```

```python
sv3.mean()
###decision tree te accuracy aro basi kome jai
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
sv4=cross_val_score(GaussianNB(),x,y,cv=5)
sv4
```

```python
sv4.mean()
###naive bayes a o accuracy kome jai
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
 sv5=cross_val_score(KNeighborsClassifier(n_neighbors=3),x,y,cv=5)
sv5
```

```python
sv5.mean()
```
