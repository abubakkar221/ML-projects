# svm ki

# svm holo ata 2ta datar modde amn akta maximum gab barkore ba line tanbe jekan theke 2ta datar durutto sob theke basi hoi

```python
x=df.drop("diagnosis",axis=1)
y=df["diagnosis"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x_scaler=mms.fit_transform(x)
x_scaler
```

```python
x=pd.DataFrame(data=x_scaler)
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41)
```

# akn amra model implement korbo

```python
from sklearn.svm import SVC
svm_model=SVC() #akane default kernel nese ,,extra kno perameter dewa hoinai svc ar modde tai.
#default:kernel="rbf",c=1,gamma="scale"
#ata dewar asol karon holo model kon pertation a kaj korbe,,maje akta dak deye naki onk gula class ar maj deye dag deye etc
svm_model.fit(x_train,y_train)
```

# svm=support vector machine
# svc=support vector classification

```python
svm_model.score(x_test,y_test)
```

# akn amra different kernel deye accuracy dakhbo

```python
#kernel sadaronto 5 doroner hoye thake jrmon kernel=linear,poly,rbf,sigmoid ,precomputed,,
###ar C ar mane holo model koto ta strict ba kothor hobe,,c ar man basidele model basi strict hobe ,,c ar man komdele model kom strict hbe,kom vul dorbe
```

```python
from sklearn.svm import SVC
model=SVC(kernel="linear", C=20,gamma="auto") #gamma auto dete hoi nijera select korar somoi
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
