```python
import numpy as np
import pandas as pd
import os
import glob


os.chdir("C:/Users/arman/ostad_code/archive")


files = glob.glob("*.csv")

print("Files found:", files)


df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

df
```

```python
df.columns
```

```python
df[" Destination Port"]
```

```python
df.columns = df.columns.str.strip() #proti ta column name ar samne pesoner space kaita dai
```

```python
df["Label"]
```

```python
df['Label'].value_counts()
```

```python
df['Label'].value_counts().shape
```

```python
def simplify_label(label):
    if label == "BENIGN":
        return "Benign"
    elif "DoS" in label or "DDoS" in label:
        return "DoS"
    elif "PortScan" in label:
        return "PortScan"
    elif "Brute" in label or "Patator" in label:
        return "Brute Force"
    elif "Bot" in label:
        return "Bot"
    elif "Web Attack" in label:
        return "Web Attack"
    else:
        return "Other"

df['Label'] = df['Label'].apply(simplify_label)
```

```python
benign = df[df['Label'] == "Benign"].sample(n=50000)
attack = df[df['Label'] != "Benign"]

df = pd.concat([benign, attack])
```

```python
df['Label'].value_counts()
```

```python
df['Label'].value_counts().shape
```

```python
df_balanced = df.groupby('Label').apply(
    lambda x: x.sample(n=min(len(x), 5000))
).reset_index(drop=True)

df = df_balanced
```

```python
df['Label'].value_counts()
```

```python
df = df[df['Label'] != "Other"]
```

```python
df['Label'].value_counts()
```

```python
x= df.drop("Label", axis=1)
y = df["Label"]
```

```python
x
```

```python
y
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
df.isna().sum().sum()
```

```python
x_train = x_train.replace([np.inf, -np.inf], np.nan)
```

```python
x_train
```

```python
x_train = x_train.fillna(x_train.mean())
```

```python
x_train.isna().sum().sum()
```

```python
x_test = x_test.replace([np.inf, -np.inf], np.nan)
x_test = x_test.fillna(x_train.mean())
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred_dt = dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print("Decision Tree Report:\n")
print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
print("Random Forest Report:\n")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.metrics import accuracy_score

print("DT:", accuracy_score(y_test, y_pred_dt))
print("RF:", accuracy_score(y_test, y_pred_rf))
```

```python
label_name = le.inverse_transform([y_pred_rf[0]])[0]
print(label_name)
```

```python
THREAT_INFO = {
    "DoS": {"severity": "High", "action": "Block IP"},
    "PortScan": {"severity": "Medium", "action": "Monitor traffic"},
    "Brute Force": {"severity": "High", "action": "Lock account"},
    "Bot": {"severity": "High", "action": "Isolate device"},
    "Web Attack": {"severity": "Medium", "action": "Filter requests"},
    "Benign": {"severity": "Low", "action": "No action"}
}
```

```python
type(THREAT_INFO)
```

```python
pred = rf.predict(x_test)[0]

label = le.inverse_transform([pred])[0]

info = THREAT_INFO.get(label, {"severity": "Unknown", "action": "Investigate"})

print("🔐 Threat Detected:", label)
print("⚠️ Severity:", info["severity"])
print("📊 Confidence: N/A")  # optional
print("🛠️ Action:", info["action"])
```
